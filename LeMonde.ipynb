```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
response = requests.get("https://www.lemonde.fr/en/")
doc = BeautifulSoup(response.text, 'html.parser')
```

```python
headers = doc.find_all('h1', attrs={'class': 'article__title'})
len(headers)
```

```python
headers2 = doc.find_all('div', attrs={'class': 'article article--headlines'})
len(headers2)
```

```python
titles = doc.find_all('p', attrs={'class': 'article__title'})
len(titles)
```

```python
for title in titles:
    print("----")
    print(title)
    
```

```python
items = doc.find_all('div', attrs={'class': 'article'})
len(items)
```

```python
for item in items:
    print("----")
    print(items)
```

```python
rows = []

for item in items:
    row = {
        'Title': item.select_one(".article__title").get_text(),
        'Subhead': item.select_one('.article__desc').get_text() if item.select_one('.article__desc') else 'N/A',
        'URL': item.select_one("a")['href'],
        #'Premium': item.select_one('a').get('aria-describedby')
        'Premium': "Yes" if  item.select_one('a').get('aria-describedby') else 'No',
        'Byline': item.select_one('.article__byline').get_text() if item.select_one('.article__byline') else 'No',
        'Article type': item.select_one('article__type').get_text() if item.select_one('article__type') else "N/A",
        'Image': item.select_one(".article__media img").get('src', None) if item.select_one(".article__media") else "No URL"
    }
    rows.append(row)
rows
```

```python
df = pd.DataFrame(rows)
df.head()
```

```python
df.to_csv("Le_Monde-Auto_update.csv", index = False)
```
